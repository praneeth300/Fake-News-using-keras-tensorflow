```python
# Import the librarys
import numpy as np
import pandas as pd
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
#Load the data
data=pd.read_csv('../input/fake-news/train.csv')
data.head()
```

```python
#shape of data
data.shape
```

```python
data.label.value_counts()
```

```python
#drop the null values we do not fill any missing values with data
df=data.dropna()
```

```python
df.reset_index(inplace=True)
```

```python
x=df.iloc[:,:-1]
y=df['label']
```

```python
y.value_counts()
```

```python
x.shape
```

```python
## One hot representation

import nltk
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
```

```python
message=x.copy()
```

```python
message.head()
```

```python
message['title'][1]
```

```python
stemmer=PorterStemmer()
lemma=WordNetLemmatizer()
```

```python
import re
```

```python
corpus=[]
for i in range(0,len(message)):
    review=re.sub("[^a-zA-Z]"," ",message['title'][i])
    review=review.lower()
    review=review.split()
    ps=[stemmer.stem(word) for word in review if not word in stopwords.words('english')]
    ps=" ".join(ps)
    corpus.append(ps)
```

```python
corpus
```

```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
voc_size=5000
```

```python
one_hot=[one_hot(words,voc_size) for words in corpus]
one_hot
```

## Embedding Representation

```python
sent_length=20
embed_doc=pad_sequences(one_hot,padding='pre',maxlen=sent_length)
embed_doc
```

```python
embed_doc[0]
```

```python
from tensorflow.keras.layers import LSTM
```

```python
#Create a moodel
dim=40
model=Sequential()
model.add(Embedding(voc_size,dim,input_length=sent_length))
model.add(LSTM(100))
model.add(Dense(1,activation='sigmoid'))
#Compile the model
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
print(model.summary())
```

```python
len(embed_doc)
```

```python
len(y)
```

```python
x_final=np.array(embed_doc)
y_final=np.array(y)
```

```python
print(x_final.shape,y_final.shape)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_final,y_final,test_size=0.3,random_state=43)
```

```python
#model training
model.fit(x_train,y_train,validation_data=(x_test,y_test),epochs=10,batch_size=64)
```

```python
y_pred=model.predict_classes(x_test)
```

```python
from sklearn.metrics import confusion_matrix

print(confusion_matrix(y_test,y_pred))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_pred))
```
